Vamos a hacer más ejercicios para practicar con las estructuras de control de flujo.

**1.**-  Crea una lista con datos numéricos aleatorios entre 0 y 99. Haz un menú usando while que muestre las siguientes opciones:

1. Añade elemento a la lista.
2. Elimina elemento de la lista.
3. Muestra la media de los datos.
4. Muestra la mediana de los datos.
5. Muestra el elemento máximo (debes recorrer la lista y comparar los elementos en vez de usar una función o método).
6. Salir.

```python
import random as rd
import numpy as np

lista = [rd.randint(0, 99) for _ in range(10)]

def añade_elemento():
    entrada = int(input("Añade el elemento que quieras añadir"))
    lista.append(entrada)
    return lista

def quita_elemento():
        entrada = int(input("Añade el elemento que quieras quitar"))
        if entrada in lista:
            lista.remove(entrada)
        else:
            print("Ese número no está en la lista.")
        return lista

def media():
    media_num = np.mean(lista)
    return media_num

def mediana():
    mediana = np.median(lista)
    return mediana

def maximo():
    max_val = lista[0]
    for numero in lista:
        if numero > max_val:
            max_val = numero
    return max_val

# Bucle del menú principal
while True:
    print("\n--- MENÚ ---")
    print(f"Lista actual: {lista}")
    print("1. Añade elemento a la lista.")
    print("2. Elimina elemento de la lista.")
    print("3. Muestra la media de los datos.")
    print("4. Muestra la mediana de los datos.")
    print("5. Muestra el elemento máximo.")
    print("6. Salir.")
    
    opcion = input("Elige una opción (1-6): ")

    if opcion == "1":
        añade_elemento()
    elif opcion == "2":
        quita_elemento()
    elif opcion == "3":
        print(f"La media es: {media()}")
    elif opcion == "4":
        print(f"La mediana es: {mediana()}")
    elif opcion == "5":
        print(f"El máximo es: {maximo()}")
    elif opcion == "6":
        print("Saliendo del programa...")
        break
    else:
        print("Opción no válida, intenta de nuevo.")
```
